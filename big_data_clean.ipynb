```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv ("/home/paola/Descargas/spotify.csv", encoding='latin-1')
df.loc[300:310]
```

```python
#ELIMINAR TODOS LOS REGISTROS DE TRACK_NAME QUE CONTENGAN "½"
mask = df['track_name'].str.contains('½')
df.drop(df[mask].index, inplace=True)
#Ajustar el index
df.reset_index(drop=True, inplace=True)
df.loc[285:295, 'track_name':'artist(s)_name']
```

```python
#ELIMINAR TODOS LOS REGISTROS DE TRACK_NAME QUE CONTENGAN "½"
mask = df['artist(s)_name'].str.contains('½')
df.drop(df[mask].index, inplace=True)
#Ajustar el index
df.reset_index(drop=True, inplace=True)
df.loc[275:285, 'track_name':'artist(s)_name']
```

```python
#ELIMINAR TODOS LOS REGISTROS DE TRACK_NAME QUE CONTENGAN "ï"
mask = df['track_name'].str.contains('ï')
df.drop(df[mask].index, inplace=True)
#Ajustar el index
df.reset_index(drop=True, inplace=True)
df.loc[150:160, 'track_name':'artist(s)_name']
```

```python
#ELIMINAR TODOS LOS REGISTROS DE TRACK_NAME QUE CONTENGAN "ï"
mask = df['artist(s)_name'].str.contains('ï')
df.drop(df[mask].index, inplace=True)
#Ajustar el index
df.reset_index(drop=True, inplace=True)
df.loc[750:760, 'track_name':'artist(s)_name']
```

```python
#ELIMINAR TODOS LOS REGISTROS DE TRACK_NAME QUE CONTENGAN "ý"
mask = df['track_name'].str.contains('ý')
df.drop(df[mask].index, inplace=True)
#Ajustar el index
df.reset_index(drop=True, inplace=True)
df.loc[540:550, 'track_name':'artist(s)_name']
```

```python
#ELIMINAR TODOS LOS REGISTROS DE TRACK_NAME QUE CONTENGAN "ý"
mask = df['artist(s)_name'].str.contains('ý')
df.drop(df[mask].index, inplace=True)
#Ajustar el index
df.reset_index(drop=True, inplace=True)
df.loc[490:500, 'track_name':'artist(s)_name']
```

```python
#Buscar valores unicos en la columna de track_name y hacerlos lista para agregarle un identificador unico
df_unique_name = df['track_name'].nunique()
df_unique_name = df['track_name'].tolist()
df_unique_name_id = { valor: i for i , valor in
    enumerate(df_unique_name)}
df_unique_name_id
```

```python
#cambiar track_name por la columna con los id
df['track_name_id'] = df['track_name'].map(df_unique_name_id)
df = df.drop(['track_name'], axis=1)
df
```

```python
#eliminar las columnas que hacen ruido
df = df.drop(['artist(s)_name','key'], axis=1)
df[["id", "capital"]]
```

```python
#renombrar y tranformar las columnas de % 
df.rename(columns={'danceability_%':'danceability','valence_%':'valence','energy_%':'energy','acousticness_%':'acousticness',
                   'instrumentalness_%':'instrumentalness','liveness_%':'liveness','speechiness_%':'speechiness'}, inplace=True)
df_columns = df[['danceability','valence','energy','acousticness','instrumentalness','liveness','speechiness']]
df_columns
for i in df_columns:
    df[i] = df[i]/100.0
df
```

```python
#hacer binario los valores de mode
df['mode'] = df['mode'].replace({'Major':1, 'Minor':0})
df
```

```python
df['danceability'] = df['danceability'].apply(lambda x:0 if x <=0.67 else 1)
df
```

```python
df.info()
```

```python
df['in_deezer_playlists'] = df['in_deezer_playlists'].str.replace(',','')
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype('int64')
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].str.replace(',','')
df['in_shazam_charts'] = df['in_shazam_charts'].fillna(0)
df['in_shazam_charts'] = df['in_shazam_charts'].astype('int64')
```

```python
df['streams'] = df['streams'].str.replace('BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3','0')
df['streams'] = df['streams'].astype('int64')
df.info()
```

```python
#Exportar archivo limpio
df.to_csv('/home/paola/Descargas/spotify_limpio.csv', index=False)
```
